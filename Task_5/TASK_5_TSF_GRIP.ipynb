### Visualizing the data

```python
plt.figure(figsize=(10,5))
sns.heatmap(dataset.corr(),cbar=True,fmt =' .2f', annot=True, cmap='coolwarm')
plt.show();
```

**Interpretation:**

1.As we are dealing with the multiple independent variables(features) so It may possible that these regressor (independent) features are correlated to each other;i.e significant multicollinearity.

2.Visualizing the correlation matrix is one way to detect the multicollinearity among the regressors.

3.Here we can see that there is no significant high correlation between any two regressors. So we can say that there is no existance of multicollinarity among the regressors.

4.Good correlation can be observed between sales and profit , implying higher sales will give us higher profit.

5.Negative correlation between profit and Discount indicates that Discounts are not good for profit.

6.Correlation between Sales and Discount is negative ,indicating Higher the sale amount lower will be discount offered on it.

```python
plt.figure(figsize=(9,6))
sns.countplot(dataset['Ship Mode'])
plt.title("Visualizing count of all classes in Ship Mode",size=12,color="blue");
plt.show();
```

```python

plt.figure(figsize = (6,6))
region = dataset['Region'].value_counts()
colors = ('skyBlue' , 'red' , 'pink' , 'lightgreen')
label = dataset['Region'].unique()

plt.pie(region, 
       autopct = '%1.1f%%',
       explode = (0.08,0.08,0.08,0.08),
       shadow = True,
       colors = colors,
       labels = label);
```

**Interpretation:**

We can easily visualize that the count for South region is high as compare to other regions

```python
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Profit', data = dataset, color = 'r', label= 'Discount')
plt.legend()
```

**Interpretation:**

This graph tells us that discount of 0.2% is tolerable. but as we give higher discount than 0.2% then profit margin will decrease. So we must need to find some tradeoff between profit and discount.

```python
plt.figure(figsize = (10,4))
sns.lineplot('Discount', 'Quantity', data = dataset, color = 'b', label= 'Quantity Purchased by customer')
plt.legend()
```

If we talk about the 'Quantity Purchased' and 'Discount Offered', then it is clearly visible that this relation is well maintained by 'SuperStoreShop'
